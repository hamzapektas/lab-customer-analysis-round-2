```python
def response_rate(x):
    if x == 'No':
        return 0
    else:
        return 1
```

```python
marcusan['response rate']= marcusan['Response'].apply(lambda x: response_rate(x))
marcusan.head()
```

```python
sns.barplot(x='response rate', y='Sales Channel', data=marcusan )
plt.show()
```
